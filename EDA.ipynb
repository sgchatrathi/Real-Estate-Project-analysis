## EDA

In this Notebook, we are cleaning the King County Data set information. In this exercise, we decided to drop variables that didn't have as much use value to our stakeholder. The columns we decided to drop were latitude, longitude, waterfront, view, The square footage of interior housing living space for the nearest 15 neighbors, and  The square footage of the land lots of the nearest 15 neighbors. 

```python
import pandas as pd

df = pd.read_csv('data/kc_house_data.csv')
df
```

454 of the value in the sqft_basement column were coded as a ?. Those entries were converted to 0 as more than half of the homes in our dataset were also coded 0.

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
df_columns_clean = df.drop(['lat', 'long',
                          'waterfront','view',
                          'sqft_living15', 'sqft_lot15', 'yr_renovated']
                         ,axis='columns')
df_columns_clean

```

```python
df_columns_clean.date.sort_values(ascending=True)
```

```python
df_columns_clean
```

```python
df_columns_clean.info()
```

We also decided that we wanted to clean the Grade column as we thought the string and the integer will be easier to analyze in our model as an integer rather than a string. 

```python
df_columns_clean['grade']
```

```python
df_columns_clean['grade'].value_counts()
```

Here we used list comprehension for our new data frame called df_columns_clean to covert grade to an integer value.

```python
df_columns_clean['grade'][0].split()[0]
```

```python
df_columns_clean['grade_int'] = [val.split()[0] for val in df_columns_clean['grade']]
df_columns_clean.head()
df_columns_clean['grade_int'].astype(int)
```

```python
df_columns_clean.head()
```

Now that we have a more digestible version of our grade level, we can drop the original column represented by the string combination called grade. 

```python
df_columns_clean = df_columns_clean.drop(['grade'], axis = 'columns')
df_columns_clean
```

```python
df_columns_clean['sqft_basement'] = ['0.0' if val == '?' 
                                     else val for val in df_columns_clean['sqft_basement']]
df_columns_clean['sqft_basement'] = df_columns_clean['sqft_basement'].astype('float')
```

# Dummying Condition

We decided to create a dummy variable related to our overall condition of the house. 

```python
df_columns_clean['condition'].value_counts()
```

```python
df_columns_clean = pd.get_dummies(df_columns_clean, columns=['condition'], drop_first=True)
df_columns_clean
```

We used the zipcode information from the king county map, to bin the zipcode data into two categories of Urban and Not Urban.
https://aqua.kingcounty.gov/gis/web/VMC/boundaries/zipcodes/zipcodes.pdf  

```python
df_columns_clean['zip_category'] = ["Urban" if val >=98100 else "Not Urban" for val in df_columns_clean['zipcode']]
```

```python
df_columns_clean.head()
```

```python
df_columns_clean = df_columns_clean.drop(['zipcode'], axis = 'columns')
```

```python
df_columns_clean = pd.get_dummies(df_columns_clean, columns=['zip_category'], drop_first=True)
```

```python
df_columns_clean
```

```python
df_columns_clean = pd.get_dummies(df_columns_clean, columns=['grade_int'], drop_first=True)
```
